```python
#DRAWING A PICTURE/GRAPH INSIDE ANOTHER GRAPH ... (Layers)
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])     #big axes...control (left, right, width, height) of the canvas (from 0 to 1)
axes2 = fig.add_axes([0.2, 0.41, 0.4, 0.3])     #small canvas

axes1.plot(x, y, 'r')
axes2.plot(y, x, 'b')
```

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4 , figsize=(16,4))
axes[0].set_title('Scatter plot')
axes[0].scatter(x, x+0.25*np.random.randn(len(x)))

axes[1].set_title('Step')
axes[1].step(n, n**2, lw=2)   #lw=line width

axes[2].set_title('Bar')
axes[2].bar(n, n**2, align  ="center", width =0.5, alpha =0.6)

axes[3].set_title('fill_between')
axes[3].fill_between(x, x**2,x**3, color="green", alpha =0.6)
```

```python
"""Draw a radar chart
"""

fig = plt.figure(figsize=(6,6))
ax= fig.add_axes([ 0.0, 0.0, 0.6, 0.6], polar = True)        #[0.0, 0, 0.6, 0.6]   ---> size
t = np.linspace(0, 2*np.pi, 100)             # (class, 100 points)
ax.plot(t,t*2, color="blue", lw =3)
```

```python
"""DRAW A HISTOGRAM"""

n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize= (12, 4))

axes[0].set_title("Defalut Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed Histogram")
axes[1].hist(n, cumulative=True , bins =10)
```

```python
#Draw a contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm    #cm=color map
import matplotlib.pyplot as plt          #py plot

#creating data
delta = 0.025
x = np.arange(-3.0, 3.0 , delta)
y = np.arange(-2.0, 2.0 , delta)
X , Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2- Y**2)
Z2 = np.exp(-(X - 1)**2- (Y-1)**2)
Z= (Z1 - Z2) *2                #height

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()
CS= ax.contour(X,Y,Z)
ax.clabel(CS, inline=1, fontsize =10)
ax.set_title("CONTOUR MAP")
```
